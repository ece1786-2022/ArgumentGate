```python
import pandas as pd
```

```python
data = pd.read_csv('data_complete.csv',usecols=["data_id","delta","upvotes","statement","argument"])
```

```python
data = data.drop(data[data.argument=="deleted"].index)
```

```python
data = data.drop(data[data.argument=="[removed]"].index)
```

```python
data['sl'] = data['statement'].apply(lambda x: len(str(x).split(' ')))
```

```python
data = data.drop(data[data.sl<3].index)
```

```python
data['al'] = data['argument'].apply(lambda x: len(str(x).split(' ')))
```

```python
data = data.drop(data[data.al<10].index)
```

```python
data['label'] = 0
```

```python
changer = data[data.delta>0]
changer['label'] = 3
changer
```

```python
excellent = data[(data.delta==0) & (data.upvotes>=30)]
excellent['label'] = 2
excellent
```

```python
adaquate = data[(data.delta==0) & (data.upvotes<15) & (data.upvotes>=10)]
adaquate['label'] = 1
adaquate
```

```python
inadaquate = data[(data.delta==0) & (data.upvotes<2) & (data.upvotes>=1)]
inadaquate['label'] = 0
inadaquate = inadaquate.sample(50000)
inadaquate
```

```python
combined = pd.concat([changer,excellent,adaquate,inadaquate])
combined
```

```python
combined['input'] = "<S_START> "+combined['statement']+" <S_END <A_START> "+combined['argument']+" <A_END>"
```

```python
combined
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(combined,stratify=combined.label,train_size=0.8)
```

```python
train.to_csv('data_train_exp.csv',columns=['input','label'],index=False)
```

```python
test,validation = train_test_split(test,stratify=test.label,train_size=0.7)
```

```python
test.to_csv('data_test_exp.csv',columns=['input','label'],index=False)
```

```python
validation.to_csv('data_validation.csv',columns=['input','label'],index=False)
```
